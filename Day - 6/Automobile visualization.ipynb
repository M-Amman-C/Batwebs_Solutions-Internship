**Data Visualization**

```python
import matplotlib.pyplot as plt #importing matplotlib
```

Comparing overall horsepower trend of cars over the years

```python
plt.plot(df['model_year'],df['horsepower'])
plt.xlabel('Model_year')
plt.ylabel('Horsepower')
plt.grid()
```

Comparing overall horsepower trend of cars over the years based on country

```python
df0 = df[df.origin == "usa"]
df1 = df[df.origin == "europe"]
df2 = df[df.origin == "japan"]

plt.plot(df0['model_year'],df0['horsepower'], label="usa")
plt.plot(df1['model_year'],df1['horsepower'], label="europe")
plt.plot(df2['model_year'],df2['horsepower'], label="japan")
plt.legend()
plt.grid()
plt.xlabel('Model_year')
plt.ylabel('Horsepower')
```

Scatterplot of weight against Horsepower based on countries

```python
plt.scatter(df0['weight'], df0['horsepower'], label="usa")
plt.scatter(df1['weight'], df1['horsepower'], label="europe")
plt.scatter(df2['weight'], df2['horsepower'], label="japan")
plt.legend()
plt.show()
```

Scatterplot of Acceleration against Milage per Gallon based on countries

```python
plt.scatter(df0['acceleration'], df0['mpg'], label="usa")
plt.scatter(df1['acceleration'], df1['mpg'], label="europe")
plt.scatter(df2['acceleration'], df2['mpg'], label="japan")
plt.xlabel('Acceleration')
plt.ylabel('Milage per Gallon')
plt.legend()
plt.show()
```

Plotting graph of No.of Cylinders vs Displacement

```python
plt.plot(df0['cylinders'], df0['displacement'], label="usa")
plt.plot(df1['cylinders'], df1['displacement'], label="europe")
plt.plot(df2['cylinders'], df2['displacement'], label="japan")
plt.xlabel('Cylinders')
plt.ylabel('displacement')
plt.xlim(4,5)
plt.ylim(100,150)
plt.legend()
plt.show()
```

Classification of No. of models by country

```python
count = [df0.shape[0], df1.shape[0], df2.shape[0]]
plt.bar( [0,1,2], [df0.shape[0], df1.shape[0], df2.shape[0]] )
plt.xlabel('countries')
plt.ylabel('No. of models')
plt.show()
```

Pie chart of No. of Models grouped by Country of Origin

```python
count = [df0.shape[0], df1.shape[0], df2.shape[0]]
plt.pie([df0.shape[0], df1.shape[0], df2.shape[0]], labels=["USA","Europe","Japan"])
plt.show()
```
